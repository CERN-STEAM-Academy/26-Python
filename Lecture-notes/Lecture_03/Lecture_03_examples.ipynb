## Activation functions as elementwise maps

```python
import numpy as np

z = np.array([-3.0, -1.0, 0.0, 1.0, 3.0])

a_tanh = np.tanh(z)
a_sigmoid = 1.0 / (1.0 + np.exp(-z))
a_relu = np.maximum(0.0, z)

print("z:       ", z)
print("tanh:    ", np.round(a_tanh, 3))
print("sigmoid: ", np.round(a_sigmoid, 3))
print("ReLU:    ", a_relu)
```

```python
import numpy as np
import matplotlib.pyplot as plt

z = np.linspace(-5, 5, 200)

plt.plot(z, np.tanh(z), label="tanh")
plt.plot(z, 1 / (1 + np.exp(-z)), label="sigmoid")
plt.plot(z, np.maximum(0, z), label="ReLU")

plt.axhline(0, linewidth=0.2)
plt.axvline(0, linewidth=0.2)
plt.legend()
plt.xlabel("z")
plt.ylabel("activation(z)")
plt.show()
```

## A minimal automatic differentiation example

```python
import numpy as np
import torch

# Simple artificial data: y = 2x + 1
x = np.linspace(-1.0, 1.0, 20)
y = 2.0 * x + 1.0

x_t = torch.tensor(x, dtype=torch.float32)
y_t = torch.tensor(y, dtype=torch.float32)

w = torch.tensor(0.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)

y_pred = w * x_t + b
loss = 0.5 * torch.mean((y_pred - y_t)**2)

loss.backward()

print("loss:", loss.item())
print("w.grad:", w.grad.item())
print("b.grad:", b.grad.item())
```
