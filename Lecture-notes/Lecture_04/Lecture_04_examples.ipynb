## Python numbers and NumPy dtypes are not the same

```python
import numpy as np

py_value = 30_000 + 30_000
print(py_value)

a = np.array([30_000], dtype=np.int16)
print(a + a)      
print((a + a).dtype)
```

```python
import numpy as np

a = 0.1 + 0.2
b = 0.3

print(a == b)
print(np.isclose(a, b))
print(a)
print(b)
print(a - b)

info = np.finfo(np.float64)

print("\n")
print(info.eps)    
print(info.tiny)   
print(info.max)  

x = 1.0
print("\n")
print(np.nextafter(x, 2.0) - x)
print(np.nextafter(0, 1))
```

## Choosing precision deliberatel

```python
import numpy as np

x64 = np.linspace(0.0, 1.0, 1_000_000, dtype=np.float64)
x32 = x64.astype(np.float32)

print(x64.nbytes / 1024**2)
print(x32.nbytes / 1024**2)

s1 = np.sum(x32)
s2 = np.sum(x32, dtype=np.float64)

print(s1, s2)
```

## Stable formulas can matter more than more digits

```python
import numpy as np

x = 1e-12

a = np.log(1.0 + x)
b = np.log1p(x)

print(a)
print(b)

y = 1e8

bad = np.sqrt(y + 1.0) - np.sqrt(y)
good = 1.0 / (np.sqrt(y + 1.0) + np.sqrt(y))

print("\n")
print(bad)
print(good)

print("\n")
for y in [1e8, 1e12, 1e16, 1e20]:
    bad = np.sqrt(y + 1.0) - np.sqrt(y)
    good = 1.0 / (np.sqrt(y + 1.0) + np.sqrt(y))
    print(y, bad, good)
```

## Lazy versus materialized data

```python
def read_sensor_value(i):
    print(f"Reads the measurement {i}")
    return i * 0.5

values = [read_sensor_value(i) for i in range(1_000_000)]

first_large = None

for value in values:
    if value > 10:
        first_large = value
        break

print(first_large)
```

## Dataclasses for configuration

```python
from dataclasses import dataclass
from pathlib import Path

@dataclass(frozen=True)
class ExperimentConfig:
    xmin: float = -10.0
    xmax: float = 10.0
    nx: int = 1001

    diffusion: float = 0.1
    tmax: float = 5.0

    output_dir: Path = Path("results")

data = ExperimentConfig(
    xmin=0.0,
    xmax=1.0,
    nx=500,
    tmax=10.0
)

data.xmin = 10
```

## Dataclass validation

```python
from dataclasses import dataclass
@dataclass(frozen=True)
class Grid1D:
    xmin: float
    xmax: float
    nx: int
    def __post_init__(self):
        if self.nx < 2:
            raise ValueError("nx must be at least 2")
        if self.xmax <= self.xmin:
            raise ValueError("xmax must be larger than xmin")
    @property
    def dx(self):
        return (self.xmax - self.xmin) / (self.nx - 1)
    def points(self):
        return np.linspace(self.xmin, self.xmax, self.nx)

grid = Grid1D(xmin = 10, xmax = 15, nx = 100)

print(grid.dx)
```
