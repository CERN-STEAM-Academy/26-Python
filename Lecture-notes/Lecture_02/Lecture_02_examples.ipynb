## Masks as physical subdomains

```python
import numpy as np

x = np.linspace(-8.0, 8.0, 2000)

sigma = 1.2
k0 = 8.0

psi = np.exp(-(x/sigma)**2) * np.exp(1j*k0*x)
rho = np.abs(psi)**2

norm = np.trapezoid(rho, x)
x_mean = np.trapezoid(x * rho, x) / norm
width = np.sqrt(np.trapezoid((x - x_mean)**2 * rho, x) / norm)

window = np.abs(x - x_mean) < 2.0 * width
print(window)
#with np.printoptions(threshold=np.inf):
#    print(window)
```

## Shared axes and fair comparison

```python
import numpy as np
import matplotlib.pyplot as plt

x_sweep = np.linspace(-2, 2, 400)
gamma = np.linspace(0.2, 2.0, 20)

U = np.array([
    np.exp(-g * x_sweep**2) * np.cos(2 * np.pi * x_sweep)
    for g in gamma
])

fig, axes = plt.subplots(
    1, 2,
    sharex=True,
    sharey=True,
    figsize=(7.0, 3.0)
)

axes[0].plot(x_sweep, U[0])
axes[0].set_title(fr"$\gamma={gamma[0]:.2f}$")

axes[1].plot(x_sweep, U[-1])
axes[1].set_title(fr"$\gamma={gamma[-1]:.2f}$")

for ax in axes:
    ax.set_xlabel(r"$x$")
    ax.grid(True, alpha=0.3)

axes[0].set_ylabel(r"$u(x)$")

fig.tight_layout()
plt.show()
```

## Physical axes with imshow

```python
import numpy as np
import matplotlib.pyplot as plt

x2 = np.linspace(-5.0, 5.0, 300)
y2 = np.linspace(-3.0, 3.0, 180)

X2, Y2 = np.meshgrid(x2, y2, indexing="ij")

sigma_x = 1.1
sigma_y = 0.7
x0, y0 = 1.0, -0.4

rho2 = np.exp(-((X2 - x0)**2/(2*sigma_x**2) + (Y2 - y0)**2/(2*sigma_y**2)))

R2 = np.sqrt(X2**2 + Y2**2)
wave2 = np.sin(3*R2) * np.exp(-0.15*R2**2)

dx2 = x2[1] - x2[0]
dy2 = y2[1] - y2[0]

mass2 = np.sum(rho2) * dx2 * dy2
x_cm = np.sum(X2 * rho2) * dx2 * dy2 / mass2
y_cm = np.sum(Y2 * rho2) * dx2 * dy2 / mass2

imax, jmax = np.unravel_index(np.argmax(rho2), rho2.shape)
x_peak, y_peak = x2[imax], y2[jmax]
```

```python
fig, ax = plt.subplots(figsize=(5.0, 3.5))
im = ax.imshow(rho2)
fig.colorbar(im, ax=ax, label="array value")
ax.set_xlabel("column index")
ax.set_ylabel("row index")
ax.set_title("Displayed in array coordinates")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5.5, 3.5))
im = ax.imshow(
    rho2.T,
    extent=[x2.min(), x2.max(), y2.min(), y2.max()],
    origin="lower",
    aspect="auto"
)
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$y$")
fig.colorbar(im, ax=ax, label=r"$\rho(x,y)$")
plt.show()
```

```python
print("expected peak:", x_peak, y_peak)

fig, ax = plt.subplots(figsize=(5.5, 3.5))
im = ax.imshow(rho2.T, extent=[x2.min(), x2.max(), y2.min(), y2.max()], origin="lower", aspect="auto")

ax.plot(x_peak, y_peak, "o", label="maximum")
ax.plot(x_cm, y_cm, "x", markersize=8, label="centre of mass")

ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$y$")
ax.legend()
fig.colorbar(im, ax=ax, label=r"$\rho$")
plt.show()
```

## Controlling false contrast

```python
import numpy as np
import matplotlib.pyplot as plt

x2 = np.linspace(-5.0, 5.0, 300)
y2 = np.linspace(-3.0, 3.0, 180)

X2, Y2 = np.meshgrid(x2, y2, indexing="ij")

sigma_x = 1.1
sigma_y = 0.7
x0, y0 = 1.0, -0.4

rho2 = np.exp(-((X2 - x0)**2/(2*sigma_x**2) + (Y2 - y0)**2/(2*sigma_y**2)))

noisy_field = rho2.copy()
noisy_field[20, 20] = 20.0 * noisy_field.max()

vmin, vmax = np.percentile(noisy_field, [1, 99])

fig, ax = plt.subplots(figsize=(5.5, 3.5))
pcm = ax.pcolormesh(X2, Y2, noisy_field, shading="auto",vmin=vmin, vmax=vmax)

#pcm = ax.pcolormesh(X2, Y2, noisy_field, shading="auto")

ax.set_title("colour range: 1st--99th percentile")
fig.colorbar(pcm, ax=ax, label=r"$\rho$")
plt.show()
```

## Masks and thresholds on fields

```python
import numpy as np
import matplotlib.pyplot as plt

x2 = np.linspace(-5.0, 5.0, 300)
y2 = np.linspace(-3.0, 3.0, 180)

X2, Y2 = np.meshgrid(x2, y2, indexing="ij")

sigma_x = 1.1
sigma_y = 0.7
x0, y0 = 1.0, -0.4

rho2 = np.exp(-((X2 - x0)**2/(2*sigma_x**2) + (Y2 - y0)**2/(2*sigma_y**2)))

R2 = np.sqrt(X2**2 + Y2**2)
wave2 = np.sin(3*R2) * np.exp(-0.15*R2**2)

mask = rho2 > 0.5 * rho2.max()
masked_wave = np.ma.array(wave2, mask=~mask)

fig, ax = plt.subplots(figsize=(5.5, 3.5))
pcm = ax.pcolormesh(X2, Y2, masked_wave, shading="auto")

ax.contour(X2, Y2, mask, levels=[0.5], linewidths=1.5)
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$y$")
ax.set_aspect("equal")
fig.colorbar(pcm, ax=ax, label=r"$w$ inside selected region")
plt.show()
```

## Plotting a residual map

```python
import numpy as np
import matplotlib.pyplot as plt

x2 = np.linspace(-5.0, 5.0, 300)
y2 = np.linspace(-3.0, 3.0, 180)

X2, Y2 = np.meshgrid(x2, y2, indexing="ij")

k, ell = 2.0, 3.0

u2 = np.sin(k*X2) * np.sin(ell*Y2)

uxx = (u2[2:, 1:-1] - 2*u2[1:-1, 1:-1] + u2[:-2, 1:-1]) / dx2**2
uyy = (u2[1:-1, 2:] - 2*u2[1:-1, 1:-1] + u2[1:-1, :-2]) / dy2**2

R_pde = uxx + uyy + (k**2 + ell**2) * u2[1:-1, 1:-1]

Xi = X2[1:-1, 1:-1]
Yi = Y2[1:-1, 1:-1]

fig, ax = plt.subplots(figsize=(5.5, 3.5))
lim = np.max(np.abs(R_pde))
pcm = ax.pcolormesh(Xi, Yi, R_pde, shading="auto",
cmap="RdBu_r", vmin=-lim, vmax=lim)

ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$y$")
ax.set_title(fr"$\|R\|_\infty = {lim:.2e}$")
fig.colorbar(pcm, ax=ax, label="residual")
plt.show()
```

## Vector fields with quiver

```python
import numpy as np
import matplotlib.pyplot as plt

xv = np.linspace(-2.0, 2.0, 25)
yv = np.linspace(-2.0, 2.0, 25)
Xv, Yv = np.meshgrid(xv, yv, indexing="ij")

Vx = -Yv
Vy = Xv

fig, ax = plt.subplots(figsize=(4.5, 4.0))
ax.quiver(Xv, Yv, Vx, Vy)

ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$y$")
ax.set_aspect("equal")
```
