## One-dimensional arrays are not automatically column vectors

```python
import numpy as np

x = np.array([1.0, 2.0, 3.0])
A = np.array([
    [10.0, 20.0, 30.0],
    [40.0, 50.0, 60.0],
    ])
B = np.array([10.0, 20.0])[:,None]

x_row = x[None, :]

print("x.shape     =", x.shape)
print("x_row.shape =", x_row.shape)
print("A.shape     =", A.shape)

print("\nx @ A.T:")
C1 = x @ A.T
print(C1.shape)
print(C1)

print("\nx_row @ A.T:")
C2 = x_row @ A.T
print(C2.shape)
print(C2)

print('\nB @ C2')
print(B @ C2)
print("\nB @ C1:")
try:
    print(B @ C1)
except ValueError as e:
    print("Error:", e)
```

```python
import numpy as np

x = np.array([1.0, 2.0, 3.0])

print(x.shape)
print(x[None, :].shape)

print((x @ x).shape)
print((x[None, :] @ x[:, None]).shape)
```

## Grids and sampled fields

```python
import numpy as np

x = np.array([10, 20, 30])
y = np.array([1, 2])

xv, yv = np.meshgrid(x, y, indexing="ij")

print("indexing='ij'")
print("xv:")
print(xv)
print("yv:")
print(yv)

print()

xv, yv = np.meshgrid(x, y, indexing="xy")

print("indexing='xy'")
print("xv:")
print(xv)
print("yv:")
print(yv)
```

## Einstein summation notation in NumPy

$$
A =
\begin{bmatrix}
1 & 2\\
3 & 4
\end{bmatrix},
\qquad
x =
\begin{bmatrix}
5\\
6
\end{bmatrix}
$$

$$
y_i = \sum_j A_{ij}x_j
$$

$$
y =
\begin{bmatrix}
1 & 2\\
3 & 4
\end{bmatrix}
\begin{bmatrix}
5\\
6
\end{bmatrix}
=
\begin{bmatrix}
17\\
39
\end{bmatrix}
$$

$$
\mathrm{dot} = \sum_i x_i y_i
$$

$$
\mathrm{dot} =
\begin{bmatrix}
5 & 6
\end{bmatrix}
\begin{bmatrix}
17\\
39
\end{bmatrix}
=
5\cdot17 + 6\cdot39
=
319
$$

$$
\mathrm{dot}
=
x^T A x
=
\begin{bmatrix}
5 & 6
\end{bmatrix}
\begin{bmatrix}
1 & 2\\
3 & 4
\end{bmatrix}
\begin{bmatrix}
5\\
6
\end{bmatrix}
=
319
$$

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])
x = np.array([5, 6])

y = np.einsum("ij,j->i", A, x)
dot = np.einsum("i,i->", x, y)

print(y)
print(dot)
```

```python
v = np.array([1.0, 2.0])
M = np.array([[2.0, 1.0],
              [1.0, 3.0]])

energy = np.einsum("i,ij,j->", v, M, v)
print(energy)
```

```python
energy = np.einsum("...i,ij,...j->...", v, M, v)
print(energy)
```

## Boundary conditions are part of the array model

## Views, copies, and memory awareness

```python
import numpy as np

u = np.arange(10.0)
print(u)
v = u[2:7]
v[0] = -1.0
print(u)
```

```python
u = np.arange(10.0)
print(u)
w = 2.0*u + 1.0
print(u)
print(w)
```
